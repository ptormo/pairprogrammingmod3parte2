```python

import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

```

```python
df = pd.read_csv("world_data_full_apply_nulos.csv", index_col=0)
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.columns
```

En los barplot las barras siempre tienen que estar ordenadas de mayor a menor.

En los barplot y countplot las barras tienen que tener un color diferente a los que vienen por defecto.

```python
#¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?

sns.histplot(data=df, x='birth_rate', bins=15, kde=True, color='violet')
plt.title('distribución tasa de natalidad')
plt.xlabel('natalidad')
plt.ylabel('frecuencia') #son los paises, es decir, hay muchos paises con un conteo de natalidad bajo, aprox 10, y pocos paises con un conteo de natalidad alto
```

```python
#¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? 
# Proporciona una breve descripción de los patrones que observes.

sns.scatterplot(data=df, x='infant_mortality', y='life_expectancy', size=2, color='violet')
plt.title('Relación mortalidad infantil/esperanza de vida')
plt.xlabel('mortalidad infantil')
plt.ylabel('esperanza de vida')

#Cuanta más mortalidad infantil hay, menor esperanza de vida, o lo que es lo mismo, mayor esperanza de vida a menor mortalidad infantil
```

```python
sns.boxplot(data=df, x='birth_rate', color='green')
plt.title('distribución tasa de natalidad')
plt.xlabel('natalidad')
plt.ylabel('frecuencia')
```

```python
sns.regplot(data=df, x='infant_mortality', y='life_expectancy', color='violet')
plt.title('Relación mortalidad infantil/esperanza de vida')
plt.xlabel('mortalidad infantil')
plt.ylabel('esperanza de vida')
```
